# Importing CPD and Bayesian Model from pgmpy

```python
from pgmpy.factors.discrete import TabularCPD
from pgmpy.models import BayesianModel
```

#Defining the Bayesian Network

```python
# adding the nodes in a directional way
question_model = BayesianModel([('A','B'), 
                                ('C', 'D'), 
                                ('B', 'E'), 
                                ('B', 'F'), 
                                ('D', 'F'), 
                                ('E', 'H'), 
                                ('E', 'I'), 
                                ('F', 'G')])
```

#Defining Conditional Probability Distribution Tables

```python
# Node A
a_cpd = TabularCPD(variable = 'A',
                          variable_card = 2,
                          values = [[.1], [.9]])
```

```python
# Node C
c_cpd = TabularCPD(
                          variable = 'C',
                          variable_card = 2,
                          values = [[.5], [.5]])
```

```python
# Node H
h_cpd = TabularCPD(
                          variable = 'H',
                          variable_card = 2,
                          values = [[.5, .7],[.5, .3]],
                          evidence = ['E'],
                          evidence_card = [2])
```

```python
# Node I
i_cpd = TabularCPD(
                          variable = 'I',
                          variable_card = 2,
                          values = [[.8, .7],[.2, .3]],
                          evidence = ['E'],
                          evidence_card = [2])
```

```python
# Node B
b_cpd = TabularCPD(
                          variable = 'B',
                          variable_card = 2,
                          values = [[.4, .9],[.6, .1]],
                          evidence = ['A'],
                          evidence_card = [2])
```

```python
# Node D
d_cpd = TabularCPD(
                          variable = 'D',
                          variable_card = 2,
                          values = [[.3, .6],[.7, .4]],
                          evidence = ['C'],
                          evidence_card = [2])
```

```python
# Node E
e_cpd = TabularCPD(
                          variable = 'E',
                          variable_card = 2,
                          values = [[.1, .2],[.9, .8]],
                          evidence = ['B'],
                          evidence_card = [2])
```

```python
# Node G
g_cpd = TabularCPD(
                          variable = 'G',
                          variable_card = 2,
                          values = [[.9, .1],[.1, .9]],
                          evidence = ['F'],
                          evidence_card = [2])
```

```python
# Node F
f_cpd = TabularCPD(variable = 'F',
                                variable_card = 2,
                                values = [[.7, .4, .2, 0], [.3, .6, .8, 1]],
                                evidence = ['B', 'D'],
                                evidence_card = [2, 2])
```

```python
# Adding all the CPDs to the Bayesian Model
question_model.add_cpds (a_cpd, b_cpd, c_cpd, d_cpd, e_cpd, f_cpd, g_cpd, h_cpd, i_cpd)
```

```python
# Showing all the nodes present in the Bayesian Model
question_model.nodes()
```

# Making Inferences

```python
from pgmpy.inference import VariableElimination
```

```python
question_infer = VariableElimination(question_model)
```

```python
prob_I = question_infer.query(variables = ['I'])
print(prob_I)
```

# P (I | C)

```python
prob_I_C = question_infer.query(variables = ['I'], evidence = {'C':0})
print(prob_I_C)
```

# P (E | ~D)

```python
prob_E_notD = question_infer.query(variables = ['E'], evidence = {'D':1})
print(prob_E_notD)
```

# P (G | B, ~D)

```python
prob_G_BandnotD = question_infer.query(variables = ['G'], evidence = {'B':0, 'D':1})
print(prob_G_BandnotD)
```
